```python
plt.plot(ages_x,dev_y,'k--',label='all devs')
plt.plot(ages_x,py_dev_y,'b',label='python')
plt.xlabel('ages')
plt.ylabel('salary')
plt.title('median salary by age')

plt.legend()
plt.show()
```

```python
plt.plot(ages_x,dev_y,color='k',linestyle='--',marker='.',label='all devs')
plt.plot(ages_x,py_dev_y,color='b',marker='o',label='python')
plt.xlabel('ages')
plt.ylabel('salary')
plt.title('median salary by age')

# for grid
plt.grid(True)

plt.legend()
plt.show()
```

```python

plt.plot(ages_x,py_dev_y,label='python')
plt.plot(ages_x,js_dev_y,label='javascript')

# all devs moved down because of clear visibisity (otherwise hidden under js plot)
plt.plot(ages_x,dev_y,color='#444444',linestyle='--',linewidth=3,label='all devs')
plt.xlabel('ages')
plt.ylabel('salary')
plt.title('median salary by age')

plt.legend()

# padding is done by tight_layout
plt.tight_layout()

plt.savefig('plot.png')

plt.show()
```
